```python
x=df_train[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','isCabin','Selected_Title']]
y=df_train['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=19)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(x_train)
```

```python
x_trains=scaler.transform(x_train)
x_tests=scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
dict1={'n_estimators':[50,75,100,150,200,300],
       'max_depth':[1,2,3,4,5],
       'min_samples_leaf':[1,2,3,4,5],
       'min_samples_split':[2,3,4,5,6]}
dict2={'n_estimators':[10,25,50,75,100,150,200,300],
       'max_depth':[1,2,3,4,5],
       'min_samples_leaf':[1,2,3,4,5],
       'min_samples_split':[2,3,4,5,6]}
```

```python
random_cv= RandomizedSearchCV(RandomForestClassifier(random_state=17), param_distributions=dict2,cv=5,verbose=2)
```

```python
random_cv.fit(x,y)
```

```python
random_cv.best_score_
```

```python
yp=random_cv.predict(df_test[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','isCabin','Selected_Title']])
```

```python
df_submission = pd.DataFrame({'PassengerId':df_test['PassengerId'], 'Survived':yp})
```

```python
df_submission.to_csv('rf_cv.csv', index=False)
```

```python
random_cv.best_params_
```
